```python
from pyspark.sql.functions import col, window, avg, min, max, count, sum, when
```

```python
spark.sql("USE CATALOG hive_metastore")
spark.sql("USE traffic_db") 
```

```python
silver_df = spark.table("traffic_silver")  

display(silver_df.limit(100))
```

```python
from pyspark.sql.functions import *
from pyspark.sql import functions as F

gold_hourly_df = (
    silver_df
    .groupBy(
        col("signal"),
        window(col("event_time"), "1 hour").alias("time_window")
    )
    .agg(
        count("*").alias("total_records"),

        avg("speed").alias("avg_speed"),
        min("speed").alias("min_speed"),
        max("speed").alias("max_speed"),

        countDistinct("vehicle_id").alias("unique_vehicles"),

        sum(when(col("signal") == "RED", 1).otherwise(0)).alias("red_count"),
        sum(when(col("signal") == "YELLOW", 1).otherwise(0)).alias("yellow_count"),
        sum(when(col("signal") == "GREEN", 1).otherwise(0)).alias("green_count"),

        avg(when(col("signal") == "RED", col("speed"))).alias("avg_speed_on_red"),
        avg(when(col("signal") == "GREEN", col("speed"))).alias("avg_speed_on_green"),
    )
    .select(
        col("signal"),
        col("time_window.start").alias("window_start"),
        col("time_window.end").alias("window_end"),

        "total_records",
        "avg_speed", "min_speed", "max_speed",
        "unique_vehicles",
        "red_count", "yellow_count", "green_count",
        "avg_speed_on_red", "avg_speed_on_green"
    )
)
gold_hourly_df = gold_hourly_df.drop("time_window")

# remove duplicate window columns after joins
for c in gold_hourly_df.columns:
    if c.startswith("window."):
        gold_hourly_df = gold_hourly_df.drop(c)

display(gold_hourly_df.limit(120))
```

```python
gold_path = "s3a://subha-trafficdata-project/gold/transformed_aggreagated/"
```

```python
(
    gold_hourly_df
    .write
    .mode("overwrite")      # later: "append" for incremental
    .format("delta")
    .save(gold_path)
)
```

```python
spark.sql("USE CATALOG hive_metastore")
spark.sql("USE traffic_db")

spark.sql(f"""
    CREATE TABLE IF NOT EXISTS traffic_gold
    USING DELTA
    LOCATION '{gold_path}'
""")
```

```python
gold_df = spark.table("traffic_db.traffic_gold")
display(gold_df.limit(50))
```

```python
import matplotlib.pyplot as plt

df = gold_df.select("signal", "avg_speed").groupBy("signal").avg("avg_speed").toPandas()

plt.figure(figsize=(6,4))
plt.bar(df["signal"], df["avg(avg_speed)"])
plt.title("Average Speed by Signal")
plt.xlabel("Signal")
plt.ylabel("Average Speed (km/h)")
plt.show()
```

```python
df = gold_df.select("window_start", "avg_speed").orderBy("window_start").toPandas()

plt.figure(figsize=(10,4))
plt.plot(df["window_start"], df["avg_speed"], marker="o")
plt.title("Congestion Trend (Average Speed Over Time)")
plt.xlabel("Time Window")
plt.ylabel("Average Speed")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
df = gold_df.select("signal", "total_records").groupBy("signal").sum("total_records").toPandas()

plt.figure(figsize=(6,4))
plt.bar(df["signal"], df["sum(total_records)"], color=["red", "green", "yellow"])
plt.title("Vehicle Events Count by Signal")
plt.xlabel("Signal")
plt.ylabel("Total Records")
plt.show()
```

```python
import seaborn as sns

df = gold_df.select("avg_speed", "min_speed", "max_speed").toPandas()

plt.figure(figsize=(7,5))
sns.boxplot(data=df)
plt.title("Speed Distribution (Min / Avg / Max)")
plt.ylabel("Speed (km/h)")
plt.show()
```

```python
df = gold_df.select(
    "window_start", 
    "red_count", 
    "avg_speed_on_red"
).orderBy("window_start").toPandas()

plt.figure(figsize=(10,4))
plt.plot(df["window_start"], df["red_count"], label="Red Light Violations", marker="o")
plt.plot(df["window_start"], df["avg_speed_on_red"], label="Avg Speed on Red", marker="x")
plt.title("Red Light Violations and Speed on Red Over Time")
plt.xlabel("Time Window")
plt.legend()
plt.xticks(rotation=40)
plt.grid()
plt.show()
```
