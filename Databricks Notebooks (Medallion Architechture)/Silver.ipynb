```python
bronze_df = spark.table("hive_metastore.traffic_db.traffic_bronze")
display(bronze_df.limit(100))
```

```python
from pyspark.sql import functions as F

silver_df = (
    bronze_df
    .withColumn("speed", F.col("speed").cast("int"))                 # convert type
    .withColumn("vehicle_id", F.col("vehicle_id").cast("int"))       # convert type
    .withColumn("event_time", F.col("ingest_ts").cast("timestamp"))  # rename timestamp
    .filter(F.col("speed") >= 0)                                     # simple filter
    .dropDuplicates(["vehicle_id", "event_time"])                    # avoid duplicates
)

display(silver_df.limit(100))
```

```python
silver_path = "s3a://subha-trafficdata-project/silver/traffic_cleaned"

(
    silver_df.write
        .mode("overwrite")
        .format("delta")
        .save(silver_path)
)
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS traffic_db.traffic_silver
USING DELTA
LOCATION 's3a://subha-trafficdata-project/silver/traffic_cleaned'
""")
```
