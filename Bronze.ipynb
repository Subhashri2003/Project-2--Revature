```python
raw_path = "s3a://subha-trafficdata-project/kinesis/traffic//////"
raw_df = (
    spark.read
         .option("recursiveFileLookup", "true")
         .json(raw_path)
)

display(raw_df)
```

```python
from pyspark.sql.functions import current_timestamp

bronze_df = (
    raw_df.withColumn("ingest_ts", current_timestamp())
)

display(bronze_df)
```

```python
bronze_path = "s3://subha-trafficdata-project/bronze/traffic/"

(
    bronze_df
        .write
        .mode("overwrite")
        .format("delta")
        .save(bronze_path)
)
```
